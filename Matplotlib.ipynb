Lets plot with providing "y" axis

````python
## Import libraries
import numpy as np              # Import the NumPy library for numerical operations
import matplotlib.pyplot as plt # Import the Matplotlib library for plotting

## Create Arrays for x and Y
x = np.linspace(0, 1, 100)      # Create an array of 100 linearly spaced points between 0 and 1
y = [k*k for k in x]           # Create an array 'y' by calculating the square of each 'x' value
                                # This mimics the function y = x**2

# Create a plot using the 'plot' function
plt.plot(x, y)

# Add labels and a title to the plot
plt.xlabel('X-axis')            # Label for the x-axis
plt.ylabel('Y-axis')            # Label for the y-axis
plt.title('Plot of y = x^2')    # Title for the plot

# Display the plot
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Create a list of temperature data
Temperature = [36, 32, 30, 25, 28, 32, 34, 31, 30, 33]

# Determines the size of the Figure in inches
plt.figure(figsize=(6, 4), dpi=100)  # Set the figure size to 6 inches by 4 inches with 100 dpi

# Add a dictionary for custom title properties
dictVal = {'fontsize': 16,          # Font size of the title
           'fontweight': 'normal',  # Font weight of the title (normal)
           'color': 'b',            # Font color (blue)
           'verticalalignment': 'baseline',        # Vertical alignment of the title
           'horizontalalignment': 'center'}        # Horizontal alignment of the title

# Plot the temperature data for the first 7 elements
plt.plot(Temperature[:7], color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12, label="Delhi", markeredgecolor="k")

# Add a title to the plot using the custom title properties
plt.title("Mean Temperature Measurements", dictVal)

# Add a legend to the plot
plt.legend()

# Display a grid in the plot
plt.grid()

# Display the plot
plt.show()
````

==========================

<div style="color:black; font-family: 'Quicksand', sans-serif;font-weight:500">


xticks are the parameters which deceides which values to be displayed as ticks ( small vertical lines on the axis of plots,which forms the basis of grid() method ). 


Note :
- Matplotlib automatically adjusts the xticks values to an optimal value, based on the size of the plot and the number of x parameters

- xticklabels labels defines the values that willp be displayed on the plots at the given values of xticks

- All text related parameters such as fontsize, rotation, font weight can be controlled while providing the xticklables

- The size of both xticks and xticklabels should be the same 

- The value of xticklabel will override the actual value of xticks(x array value) while plotting

For Eg :    

`plt.xtick([1,2,3],["Monday","Tuesday","Wednesday"])`

</div>

````python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of y values with 10000 points between 0 and 1
y = np.linspace(0, 1, 10000)

# Create an array of x values with 10000 points between 0 and 1
x = np.linspace(0, 1, 10000)

# Plot the x and y values
plt.plot(x, y)

# Define custom tick locations and labels for the x-axis
plt.xticks([0, 0.1, 0.2, 0.5, 0.7, 0.99, 1.0],      # Tick positions on the x-axis
           ["sdfsd", "asda", 0.2, 0.5, 0.9, 0.99, 1.0],  # Corresponding tick labels
           rotation=60)                             # Rotate the tick labels by 60 degrees

# Display a grid in the plot
plt.grid()

# Display the plot
plt.show()
````

---

<div style="color:black; font-family: 'Quicksand', sans-serif;font-weight:500">

matplotlib proivdes a short format notation for easily providing the line plot parameters. They are defined as follows

`fmt = '[marker][line][color]'`


```
'b'    # blue markers with default shape
'or'   # red circles
'-g'   # green solid line
'--'   # dashed line with default color
'^k:'  # black triangle_up markers connected by a dotted line
```

</div>


````python
# Define temperature data for Bangalore and Delhi
Temperaturebangalore = [36, 32, 30, 25, 28, 32, 34, 31, 30, 33]
TemperatureDelhi = [40, 39, 31, 35, 39, 37, 36, 35, 34, 38]

# Determine the size of the figure
plt.figure(figsize=(6, 4), dpi=100)  # Set the figure size to 6 inches by 4 inches with 100 dpi

# Plot temperature data for Bangalore with black triangle markers and dashed lines
plt.plot(Temperaturebangalore, '^k:', label="Bengaluru")

# Plot temperature data for Delhi with blue circle markers and solid lines
plt.plot(TemperatureDelhi, 'ob-', label="Delhi")

# Add a legend to the plot
plt.legend()

# Add a title to the plot
plt.title("Temperature Comparison: Bengaluru vs Delhi")

# Display the plot
plt.show()
````

### Basic plotting excercise
---


**Question:**

Your task is to create a plot that visualizes the sine and cosine functions in the interval from 0 to \(2\pi\). Follow the steps below to create the plot with the specified attributes:

1. Generate an array \(x\) with values ranging from 0 to \(2\pi\) to represent the \(x\) coordinates for plotting.
2. Calculate the corresponding \(y\) values for the sine and cosine functions.
3. Create a plot with the following properties:
   - Plot the sine function using dashed blue lines and circular markers.
   - Plot the cosine function using solid black lines and triangular markers.
   - Set appropriate labels for both functions.
   - Add a title to the plot.
   - Place the legend outside the plot.
   - Set x-axis tick labels at intervals of \(\pi/4\) radians.
   - generate marker every for only 10th value of x
   - Add grid

Your task is to write the code to accomplish these steps and create the desired plot.


````python
#@title Solution for Basic Plotting excercise
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate x values from 0 to 2*pi
x = np.linspace(0, 2*np.pi, 100)

# Step 2: Calculate y values for sine and cosine
y_sin = np.sin(x)
y_cos = np.cos(x)

# Step 3: Create the plot
plt.plot(x, y_sin, 'b--o', label='Sine',markevery=10)  # Dashed blue lines with circular markers for sine
plt.plot(x, y_cos, 'k-o', label='Cosine',markevery=10)  # Solid black lines with circular markers for cosine

# Set x-axis tick labels at intervals of pi/4
plt.xticks(np.arange(0, 2*np.pi + np.pi/4, np.pi/4), 
           ['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$', r'$\frac{5\pi}{4}$', r'$\frac{3\pi}{2}$', r'$\frac{7\pi}{4}$', r'$2\pi$'])

# Add title and labels
plt.title('Sine and Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')

# Place legend outside the plot
plt.legend(loc='upper right')

# add grid
plt.grid()

# Display the plot
plt.show()
````

### Bar Plot

````python
# Disable LaTeX rendering for text
plt.rcParams['text.usetex'] = False

# Define the programming languages and the corresponding number of students
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]

# Create a figure with a specified size and dpi
plt.figure(figsize=(6, 4), dpi=100)

# Create a bar plot with languages on the x-axis and student counts on the y-axis
bars = plt.bar(langs, students)

# Define a list of hatch patterns to use for filling bars
hatchList = ['o', '/', "\\", '*', 'O']

# Apply hatch patterns to the bars using a loop
for index, hatch in enumerate(hatchList):
    bars[index].set_hatch(hatch)

# Add title, x-label, and y-label to the plot
plt.title("Bar Plot")
plt.xlabel("Languages")
plt.ylabel("Students")

# Display the plot
plt.show()

````


### 2. Scatter Plot

---
**Scatter Plot Summary:**
A scatter plot is a data visualization tool used to display individual data points on a two-dimensional graph. Each point represents values for two variables, with one plotted along the x-axis and the other along the y-axis. Scatter plots are useful for identifying patterns, relationships, and distributions within the data.

**Usage in Data Science:**
Scatter plots are widely used in data science for various purposes:
- **Correlation Analysis:** Scatter plots help visualize the relationship between two continuous variables, revealing whether they are positively, negatively, or not correlated.
- **Outlier Detection:** Outliers, or unusual data points, can be easily spotted in a scatter plot, aiding in identifying anomalies.
- **Regression Analysis:** Scatter plots can be used to visualize the data before fitting a regression model, helping to understand the potential relationship between variables.

### Method - 1 : Render Each Frame Seperately 
---

In this, we will use the `pause()` function within matplotlib to generate a plot, which will make it appear like an animation 



````python
y = ecg[300:4000:10]
t = time[300:4000:10]
 
## generate Plot Object 
fig,ax = plt.subplots(figsize=(6,4),tight_layout=True,dpi=100 )
 
for frame in range(len(y)):
    ax.clear()  ## Clear the Object
    ax.set_xlabel("Time(s)")
    ax.set_ylabel("ECG in mV")
    ax.grid()
    ax.plot(t[:frame],y[:frame],'ok-',markevery=[-1])
    ax.set_ylim([-1,1.5])
    fig.canvas.draw()
    plt.pause(0.00001)
    pass;
````

### Method - 2 : Update the Line Elements(Artists Object) 
---


In this, Rather than redrawing the entire plot, we can just modify the data which is stored in the line object to get the updated plots. 






````python
y = ecg[300:4000:10]
t = time[300:4000:10]
 
## generate Plot Object 
fig,ax = plt.subplots(figsize=(6,4),tight_layout=True,dpi=100 )

ln, = ax.plot([],[],'ok-',markevery=[-1]) ### Obtain the Line Object
ax.set_xlabel("Time(s)")
ax.set_ylabel("ECG in mV")

for frame in range(len(y)):
    ln.set_data(t[:frame],y[:frame])
    ax.grid()
    ax.plot(t[:frame],y[:frame],'ok-',markevery=[-1])   ## rather than clearing the canvas, just update the line object alone. 
    # ax.set_ylim([-1,1.5])
    ax.relim()
    fig.canvas.draw()
    plt.pause(0.00001)
    pass;
````

### 3. Using FuncAnimation and blit
---

This is matplotlib's inbuilt funcitonality to animate the plots by calling this functions again and again


* `init()`   -- Function initialises the plot
* `update()` -- Function updates the artists objects to update the plot

**blit** is an intelligent way to update the graphics without much effort. Additional details about blit can be refered from this [video](https://www.youtube.com/watch?v=-6g1tD66QBA)

````python
import numpy as np
import matplotlib.pyplot as plt
# from matplotlib import mp

y = ecg[300:10000:10]
t = time[300:10000:10]


fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'ok-')


## Intialises the plot
def init():
    ln, = ax.plot([],[],'ok-',markevery=[-1]) ### Obtain the Line Object
    ax.set_xlabel("Time(s)")
    ax.set_ylabel("ECG in mV")
    ax.set_ylim([-1,1.5])
    return ln,

## Updates the plot at each frame
def update(data):
    xdata.append(data[1])
    ydata.append(data[0])
    ln.set_data(xdata, ydata)
    ax.relim()
    return ln,

ani = FuncAnimation(fig, update, frames=list(zip(y,t)),
                    init_func=init, interval = 10, blit=True)
plt.show()
````

### Additional information about `FuncAnimation`
---

**Frames Parameter**:
In the FuncAnimation function call, the frames parameter specifies an iterable that defines the data for each frame of the animation. In this example, frames=list(zip(y, t)) combines the y (ECG data) and t (time) arrays using the zip function. Each pair (y[i], t[i]) represents the data for a single frame. The animation will iterate over these pairs, where each pair represents a point to be plotted at a particular time.

**Interval Parameter**:
The interval parameter controls the time delay (in milliseconds) between successive frames in the animation. In this case, interval=10 means there will be a 10-millisecond pause between updating each frame. The smaller the interval, the faster the animation will appear, and vice versa.

````python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

# Sample x and y data
x = np.linspace(0, 10, 100)  # Sample x values
y = np.sin(x)                # Sample corresponding y values

# Creating the Animation Function
def animate(frame):
    plt.clf()  # Clear the current plot
    plt.xlim(0, 10)  # Set x-axis limits
    plt.ylim(-1.5, 1.5)   # Set y-axis limits
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Animated Line Plot')

    # Plot the data up to the current frame
    plt.plot(x[:frame], y[:frame], color='blue')

# Creating the Animation
fig = plt.figure()
animation = FuncAnimation(fig, animate, frames=len(x), interval=100)

# Displaying the Animation
plt.show()
````
