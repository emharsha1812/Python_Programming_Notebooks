Label vs. Location
The main distinction between the two methods is:

## 7. DataFrame Grouping 
----


[Additional Reference](https://pandas.pydata.org/docs/user_guide/groupby.html#groupby)

```python
df = pd.DataFrame(
    {
        "A": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Male"],
        "B": ["Data Scientist", "Data Scientist", "Software", "three", "Software", "Software", "Data Scientist", "Professor"],
        "C": abs(np.random.randn(8))*100000,
        "D": abs(np.random.randn(8))*100000,
    }
)

df
```

```python
# Grouping and then applying the sum() function to the resulting groups
# df.groupby("A")[["C", "D"]].sum()
df.columns = ["Gender", "Role", "Salary","Temp"]
df.groupby(["Gender","Role"])[["Salary"]].count()
```

```python
df
```

```python
#Grouping by multiple columns forms a hierarchical index
grouped = df.groupby(["Gender","Role"])[["Salary","Temp"]].sum()
## Access by the Hierarchy of Data
grouped
```
